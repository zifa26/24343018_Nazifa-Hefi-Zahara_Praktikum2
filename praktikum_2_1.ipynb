```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def analyze_my_image(image_path, sample_path):
    analysis_results = {}

    img = cv2.imread(image_path)
    sample = cv2.imread(sample_path)

    # 1. Dimensi & Resolusi
    height, width, channels = img.shape
    resolution = width * height

    print("=== 1. DIMENSI & RESOLUSI ===")
    print("Lebar:", width)
    print("Tinggi:", height)
    print("Jumlah Channel:", channels)
    print("Resolusi:", resolution)

    # 2. Aspect Ratio
    aspect_ratio = width / height
    print("\n=== 2. ASPECT RATIO ===")
    print("Aspect Ratio:", aspect_ratio)

    # 3. Grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("\n=== 3. GRAYSCALE ===")
    print("Ukuran RGB:", img.shape)
    print("Ukuran Grayscale:", gray.shape)

    # 4. Statistik
    print("\n=== 4. STATISTIK CITRA ===")
    print("Mean:", np.mean(img))
    print("Standar Deviasi:", np.std(img))
    print("Min:", np.min(img))
    print("Max:", np.max(img))

    # 5. Histogram
    print("\n=== 5. HISTOGRAM ===")
    colors = ('b','g','r')
    for i, col in enumerate(colors):
        hist = cv2.calcHist([img],[i],None,[256],[0,256])
        plt.plot(hist)
    plt.title("Histogram RGB")
    plt.xlabel("Intensity Value")
    plt.ylabel("Pixel Count")
    plt.show()

    # 6. Perbandingan dengan Sample
    print("\n=== 6. PERBANDINGAN DENGAN SAMPLE ===")
    print("Resolusi Citra Pribadi:", resolution)
    print("Resolusi Sample:", sample.shape[0] * sample.shape[1])

    if resolution > sample.shape[0] * sample.shape[1]:
        print("Citra pribadi memiliki resolusi lebih besar.")
    else:
        print("Citra sample memiliki resolusi lebih besar.")
```

```python
analyze_my_image("foto pribadi.jpeg", "citra sample.jpeg")
```
